```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv('claimants.csv')
df.head()
```

```python
#No Use of this column in our prediction
df.drop(['CASENUM'],inplace=True,axis=1)
df
```

```python
Target=df['ATTORNEY']
Target
```

```python
Features=df.drop(columns='ATTORNEY')
Features
```

```python
df.isnull().sum()
```

```python
#Handling missing values in large dataset
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
No_Null_Features=pd.DataFrame(imputer.fit_transform(Features), columns=Features.columns)

#We have another way of dng this for small dataset
"""NO_Null_FeaturesFeatures=Features.fillna(Features.mean())
Features"""
```

```python
No_Null_Features.isnull().sum()
```

```python
#Standardization
#Import the module
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Std_Features=sc.fit_transform(No_Null_Features)
Std_Features
```

```python
#Converting Array into DF
Std_Features=pd.DataFrame(Std_Features,columns=No_Null_Features.columns)
Std_Features
```

```python
#Splitimg the data
x_train,x_test,y_train,y_test= train_test_split(Std_Features,Target,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#GridSearch of hyperparameters
svc=SVC()
params={'C':[1,2,3,4,5],'kernel':['linear','poly','rbf','sigmoid'],'gamma':[0.1,0.3,0.5,0.6]}
grid_search= GridSearchCV(svc,params)
grid_search.fit(x_train,y_train)
grid_search.best_params_
```

```python
svc=SVC(C=2.0,kernel='linear',gamma=0.1)
```

```python
#Train the model
svc.fit(x_train,y_train)
```

```python
y_pred= svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%writefile Model_Dep_app.py
import warnings
warnings.filterwarnings('ignore')
import streamlit as st
import pickle
import pandas as pd
model=pickle.load(open('Dummy1.pkl','rb'))  # rb __read binary
def user_input_varibles(): # Defining function
    #TAKING INPUTS FRM USER
    st.sidebar.header('USER INPUT PARAMETERS')
    st.title('Model Deployment using LR Model')
    CLMSEX= st.sidebar.selectbox('Choose your Gender,M=1,F=0',[0,1])
    CLMINSUR=st.sidebar.selectbox('Detail of Insurance, Y=1,N=0',[0,1])
    SEATBELT=st.sidebar.selectbox('Detail of Seatbelt,Y=1,N=0',[0,1])
    CLMAGE=st.sidebar.number_input('Enter the Age of the Car')
    LOSS=st.sidebar.number_input('Enter the Loss')
    # Creating Dictionary
    data= {'CLMSEX':CLMSEX,'CLMINSUR':CLMINSUR,'SEATBELT':SEATBELT,'CLMAGE':CLMAGE,'LOSS':LOSS}
    #Converting Dictionary into DataFrame
    features= pd.DataFrame(data,index=[0])
    return features
df= user_input_varibles() # Calling Function
st.subheader('USER_INPUT_VARIABLES')
st.write(df)
#PREDICTION
pred_prob=model.predict_proba(df)
pred= model.predict(df)

#SUB_Headers in Page
st.subheader('PREDICTED')
# Logstic Regresion consists of probability [0] & [1] i,e both No & Yes separately
st.write('Yes' if pred_prob [0][1]>0.5 else 'No')
st.subheader('PREDICT_PROB')
st.write(pred_prob)
st.header('Developed_By_VAMC_YADAV')
```
